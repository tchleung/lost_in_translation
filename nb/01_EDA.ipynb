### Waveplot visualization 

The librosa package has a built-in function that graphs the amplitude envelope of a waveform

The following are 10 sample waveplots (one per language from the dataset) 

```python
# Multiplots of 3 languages
fig, ax = plt.subplots(3,1, figsize=(16,10))
file_paths = [
    '../data/chinese/call_to_arms_0064.wav',
    '../data/japanese/meian_5057.wav',
    '../data/russian/icemarch_1337.wav']

language_list = [
    'Chinese sample',
    'Japanese sample',
    'Russian sample']

for idx, val in enumerate(file_paths):
    data, sample_rate = librosa.load(val)
    librosa.display.waveplot(data,sr=sample_rate,ax=ax[idx])
    ax[idx].set_title(language_list[idx])
    plt.tight_layout();
```

# Spectrogram

```python
lang_lst = ['chinese','dutch','finnish','french','german','greek','hungarian','japanese','russian','spanish']

for i in range(len(lang_lst)):
    path = "../data/"
    lang_dir = os.path.join(path, lang_lst[i])
    files = os.listdir(lang_dir)
    file = os.path.join(lang_dir, files[0])

    sample_rate, audio = wavfile.read(file)
    mel = librosa.power_to_db(librosa.feature.melspectrogram(audio, sample_rate), np.max)
    plt.figure(figsize=(10, 6))
    librosa.display.specshow(mel, sr=sample_rate, x_axis="time", y_axis="mel")
    plt.title(lang_lst[i])
    plt.colorbar(format='%+2.0f dB')
    plt.show()
    plt.close()
```

```python
lang_lst = ['chinese','dutch','finnish','french','german','greek','hungarian','japanese','russian','spanish']

for i in range(len(lang_lst)):
    path = "../data/"
    lang_dir = os.path.join(path, lang_lst[i])
    files = os.listdir(lang_dir)
    file = os.path.join(lang_dir, files[0])

    sample_rate, audio = wavfile.read(file)
    mel = librosa.power_to_db(librosa.feature.mfcc(audio, sample_rate, n_mfcc=40), np.max)
    plt.figure(figsize=(10, 6))
    librosa.display.specshow(mel, sr=sample_rate, x_axis="time", y_axis="mel")
    plt.title(lang_lst[i])
    plt.colorbar(format='%+2.0f dB')
    plt.show()
    plt.close()
```
